### Размеры матриц:
#### 100x100
#### 100x100

```python
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
num_threads = [1,2,3,4,5,6,7]
times = [0.007044, 0.008447, 0.005823, 0.004401, 0.003719, 0.003283, 0.005346]
plt.plot(num_threads, times)
plt.xlabel("Number of threads")
plt.ylabel("Time of work")
plt.show()
```
