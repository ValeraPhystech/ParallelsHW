```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
arr = [8.980864,\
       5.080924,\
       1.181423,\
       0.964112,\
       0.811018]
grid = np.linspace(1, len(arr), len(arr))

plt.title("N= 1000")
plt.ylabel("Time")
plt.xlabel("Procs")
plt.plot(grid, arr)
plt.show()
```

```python
arr = [8.980864,\
       5.080924,\
       1.181423,\
       0.964112,\
       0.811018]
grid = np.linspace(1, len(arr), len(arr))

acceler = [arr[0] / x for x in arr]

plt.title("Acceleration N= 1000")
plt.ylabel("Time")
plt.xlabel("Procs")
plt.plot(grid, acceler)
plt.show()
```

```python
arr = [8.980864,\
       5.080924,\
       1.181423,\
       0.964112,\
       0.811018]
grid = np.linspace(1, len(arr), len(arr))

acceler = [arr[0] / x for x in arr]
eff = [ac / p for ac, p in zip(acceler, range(1, len(arr) + 1))]
plt.title("Efficiency N= 1000")
plt.ylabel("Time")
plt.xlabel("Procs")
plt.plot(grid, eff)
plt.show()
```
